### memo
- distanceTransform()は、注目している黒ではないピクセルから一番近い黒ピクセルまでの距離を求める関数。
- 黒ピクセルが細すぎると、距離を正しく求められないっぽい？ (input5.pngのケース)


```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

def test(filename):
    src = cv2.imread(filename)
    mono = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

    kernel = np.ones((5,5),np.uint8)
    mono = cv2.erode(mono, kernel,iterations = 3)

    plt.imshow(cv2.cvtColor(mono, cv2.COLOR_GRAY2RGB))
    plt.title('mono')
    plt.show()

    dist = cv2.distanceTransform(mono,cv2.DIST_L2, 5)
    ret, img = cv2.threshold(dist, 30, 255, cv2.THRESH_BINARY_INV)

    plt.imshow(dist)
    plt.title('dist')
    plt.show()
```
